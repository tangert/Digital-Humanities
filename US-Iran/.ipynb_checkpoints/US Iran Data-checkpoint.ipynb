### Some helper methods

```python
def parse_date(date):
    """Returns an array of integers represetning the month, date, year from a string"""
    if type(date) is str:
        ds = [int(d) for d in date.split('/')]
        year = ds[2]
        month = ds[0]
        day = ds[1]
        return datetime.date(year,month,day)
    else:
        return None
```

```python
def apply(df, col, fn):
    """convenience apply function"""
    df[col] = df[col].apply(fn)
```

```python
def normalize(df, col):
    """normalizes the values of a given column"""
    result = df[col].copy()
    max_value = df[col].max()
    min_value = df[col].min()
    result = (df[col] - min_value) / (max_value - min_value)
    return result
```

```python
def gen_plot(df,x,y,color):
    plt.plot(x, y, data=df, marker='', color=color, linewidth=1, linestyle='dashed')
    plt.legend()
```

```python
main_df.columns = ["BB","ID","og_date","start_date","end_date","title","description","references","flag","notes"]
```

```python
# parse the date columns
apply(main_df,'end_date',parse_date)
apply(main_df,'start_date',parse_date)
```

```python
main_df = main_df.drop(columns=['og_date','references','flag','notes'])
```

# Basic cleaning done!


Now that our data are more readable, let's see what kinds of questions we want to ask and data we want to look at. Let's do some basic word count, POS tagging, and sentiment analysis.

```python
shared_cols = ['BB','ID','start_date','end_date']
```

```python
def get_word_counts(inp):
    tokens = nltk.tokenize.word_tokenize(inp)
    return nltk.FreqDist(t.lower() for t in tokens)
```

```python
title_df = main_df.filter(shared_cols + ['title'], axis=1)
desc_df = main_df.filter(shared_cols + ['description'], axis=1)

title_df['title_tokens'] = main_df['title'].apply(nltk.tokenize.word_tokenize)

# title length and unique words
title_df['title_len'] = main_df['title'].apply(len)
title_df['title_len_norm'] = normalize(title_df, "title_len")

title_df['title_uniques'] = main_df['title'].apply(set).apply(len)
title_df['title_unique_norm'] = normalize(title_df, "title_uniques")

# differences
title_df['normalized_diffs'] = abs(title_df['title_len_norm'] - title_df ['title_unique_norm'])
title_df['title_word_counts'] = main_df['title'].apply(get_word_counts)

title_df['title_pos_tags'] = title_df['title_tokens'].apply(nltk.pos_tag)


desc_df['description_len'] = main_df['description'].apply(len)
desc_df['description_uniques'] = main_df['description'].apply(set).apply(len)
```

```python
# normalized
gen_plot(title_df,ids,"title_len_norm","red")
gen_plot(title_df,ids,"title_unique_norm","blue")
```

```python
# Differences between the two graphs
gen_plot(title_df,ids,"normalized_diffs","green")

#average normalized difference
average_diffs = [np.mean(list(title_df["normalized_diffs"]))]*len(ids)
plt.plot(ids, average_diffs,label='average difference')
plt.legend()
```
