## LDA

Constructing basic topic models for each text.

```python
all_stops = [w for w in STOPWORDS]
```

```python
new_stops = ['god', 'lord','allah','jesus', 'thou', 'thee', 'thy', 'unto', 'ye', 'said', 'shall', 'shalt', 'yea', 'thereof']
all_stops += new_stops
```

```python
# allWordExceptStopDist = nltk.FreqDist(w.lower() for w in allWords if w not in stopwords)    
# mostCommon= allWordDist.most_common(10).keys()
```

```python
# Helper functions
def tokenize(text):
    return [token for token in simple_preprocess(text) if token not in all_stops]

def get_cleaned_transcript(df):
    # creates a 1D list of all of the speech values from the dataframe
    speech_list = get_lines(df)
    # tokenize each transcript line
    return [tokenize(line) for line in speech_list]

def get_dictionary(cleaned_docs):
    # create a dictionary of all of the words (unique tokens) in the speech
    return Dictionary(cleaned_docs)

def create_dt_mtx(dictionary, cleaned_docs):
    return [dictionary.doc2bow(doc) for doc in cleaned_docs]
```

```python
def get_topic_model(df, num_topics, passes):
    
    # first clean
    cleaned_docs = get_cleaned_transcript(df)
    
    # get the dictionary 
    dictionary= get_dictionary(cleaned_docs)
    
    # create the document term matrix
    dt_mtx = create_dt_mtx(dictionary, cleaned_docs)
    
    # create the model
    model = LDA(dt_mtx, num_topics=num_topics, id2word = dictionary, passes=passes)
    
    return model
```

```python
otm = get_topic_model(ot_df, 10, 25)
ntm = get_topic_model(nt_df, 10, 25)
qtm = get_topic_model(q_df, 10, 25)
```

```python
def create_topic_dict(tm):
    # grab the topics
    topics = tm.show_topics(num_topics=5, num_words=5, log=False, formatted=False)
    to_export = [] 
    for t in topics:
        topic_container = {}
        topic_probs = t[1]
        for p in topic_probs:
            topic_container[p[0]] = float(p[1])
        to_export.append(topic_container)
        
    return to_export
```

```python
otops = create_topic_dict(otm)
ntops = create_topic_dict(ntm)
qtops = create_topic_dict(qtm)
```

```python
all_topics = {
    "ot": otops,
    "nt": ntops,
    "qt": qtops
}
```

```python
def export_topics(d, file):    
    with open(file, 'w') as fp:
        json.dump(d, fp)
```

```python
export_topics(all_topics, 'all_topics.json')
```

## Word2Vec

Constructing a holistic word2vec model for all of the texts put together.

```python
# Global hyper parameters
hp = {
    "size": 150, # size of the one-hot-encoded word vectors
    "window": 10, # context size
    "min_count": 2,
    "workers": 10
}
```

```python
# helper function for printing words form results of model
def get_res(res):
    return [d[0] for d in res]
```

```python
def concat_dfs(dfs):
    # Ignore the index to avoid having conflicting row numbers when accessing the df
    return pd.concat(dfs, ignore_index = True)
```

```python
# use the same preprocessed text from the LDA modelling to create the word2vec relationships
all_dfs = concat_dfs([ot_df, nt_df, q_df])
cleaned_docs = get_cleaned_transcript(all_dfs)
```

```python
w2v = gensim.models.Word2Vec(
        cleaned_docs,
        size=hp["size"],
        window=hp["window"],
        min_count=hp["min_count"],
        workers=hp["workers"])

w2v.train(cleaned_docs, total_examples=len(cleaned_docs), epochs=10)
```

```python
pca = PCA(n_components=2)

w2v_X = w2v[w2v.wv.vocab]
w2v_result = pca.fit_transform(w2v_X)
```

```python
# This contains the vectors for the whole vocabulary
wv_dict = {}

for i, word in enumerate(list(w2v.wv.vocab)):
    wv_dict[word] = [float(w2v_result[i][0]),float(w2v_result[i][1])]
```

```python
topic_vectors = {}

for tm in all_topics:
    tops = all_topics[tm]
    topic_vecs = []
    for t in tops:
        vecs = {}
        for w in t:
            vecs[w] = wv_dict[w]
        topic_vecs.append(vecs)
    topic_vectors[tm] = topic_vecs
```

```python
export_topics(topic_vectors, 'all_topics+vectors.json')
```

## Polygon construction from topics

```python
from shapely import geometry as g
from dxfwrite import DXFEngine as dxf
```

```python
polygon = g.Polygon([(0, 0), (1, 1), (1, 0)])
```

```python
polygon
```

```python
polygon_dict = {}
for t in topic_vectors:
    polygons = []
    vecs = topic_vectors[t]
    for v in vecs:
        pts = []
        for w in v:
            pt = (v[w][0], v[w][1])
            pts.append(pt)
        topic_poly = g.Polygon(pts)
        polygons.append(topic_poly)
    polygon_dict[t] = polygons
```
